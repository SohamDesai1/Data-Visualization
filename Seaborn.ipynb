```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
%matplotlib inline
```

```python
df = yf.download('AAPL', start='2022-01-01', end='2022-12-31', progress=False)
df.head()
```

```python
res = sns.lineplot(x=df.index, y=df['Close'])
res.set_title('AAPL Stock Price')
res.set_ylabel('Price')
res.set_xlabel('Date')
plt.show()
```

```python
res = sns.barplot(x=df.index, y=df['Close'],color='red')
plt.show()
```

```python
res = sns.barplot(x=df.index, y=df['Close'],palette='Blues')
plt.show()
```

```python
sns.countplot(x='Close', data=df)
sns.countplot(x='Close', data=df,palette='Set2')
plt.show()
```

```python
sns.countplot(y='Close', data=df)
plt.show()
```

```python
sns.displot(df['Close'], kde=True)
plt.show()
```

```python
sns.heatmap(df.corr(),cbar=True ,annot=True)
```

```python
sns.scatterplot(x='Close', y='Open', data=df,hue='Close')
plt.show()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='Close', palette='husl')
```
